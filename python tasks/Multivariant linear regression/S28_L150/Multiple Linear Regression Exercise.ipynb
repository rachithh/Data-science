## Import the relevant libraries

```python
import pandas as pd
import numpy as nm
import statsmodels.api as sm
import matplotlib.pyplot as ml
import seaborn as sns
sns.set()
```

## Load the data

```python
data = pd.read_csv('real_estate_price_size_year.csv')
```

```python
data
```

```python
data.describe()
```

## Create the regression

### Declare the dependent and the independent variables

```python
y=data['price']
x1=data['size']
x2=data['year']
```

### Regression

```python
x=sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
ml.scatter(x1,y,c=x2,cmap='RdYlGn_r')
yhat_year = -5.772e+06+227.7009*x1 + 2916.7853*x2
yhat_size = 1.019e+05+223.1787*x1
fig = ml.plot(x1,yhat_size,lw=2,c='Black')
fig = ml.plot(x1,yhat_year,lw=2,c='Green')
ml.xlabel('Price',fontsize=20)
ml.ylabel('Size',fontsize=20)
ml.show()
```
